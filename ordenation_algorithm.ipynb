# Selection Sort 

Algoritmo de complexidade n², tem como idéia básica colocar elemento por elemento no seu lugar correrto, começando sempre do menor número do vetor até o maior. São feitos dois laços, o primeiro pra controlar o índice do vetor que está sendo ordenado o segundo para identificar o menor elemento do vetor que seja maior que o da posição anterior e assim ao fim dessa travessia fazer o swap entre o elemento que se encontra na posição a ser ordenada e o elemento que realmente pertence aquela posição.

### Importando bibliotecas

```python
import time 
import matplotlib.pyplot as plt
import numpy as np
```

### Algoritmo Selection Sort

```python
def selection_sort(vector):

    for i in range(0, len(vector) - 1):
        less = i;
        for j in range(i+1, len(vector)):
            if vector[j] < vector[less]:
                less = j
            
        vector[i],vector[less] = vector[less],vector[i]    
    return vector
```

### Vetor desordenado

# Insertion Sort

Algoritimo de ordenação n² que tem como princípio a formação de um sub-array na esquerda onde o próximo elemento do array principal é comparado e posto na posição correta no sub-array, assim ao fim do array o sub-array estará organizado de forma crescente.

Vantagens:
<ul>
    <li> É o método a ser utilizado quando o arquivo está "quase" ordenado</li>
    <li> É um bom método quando se desejar adicionar poucos elementos em um arquivo já ordenado, pois seu custo é linear</li>
     <li> O algoritmo de ordenação por inserção é estável</li>
</ul>

Desvantagens: 
<ul>
    <li> Alto custo de movimentação de elementos no vetor </li>
</ul>

### Algoritmo Insertion Sort

```python
def insertion_sort(vector):
    for i in range(1, len(vector)):
        j = i
        while j != 0 and vector[j] < vector[j - 1]:
            vector[j], vector[j-1] = vector[j-1], vector[j]
            j = j - 1
    return vector
```

# Bubble Sort 

Algoritmo natural de ordenação, que tem como princípio a comparação com o proximo elemento da estrutura, caso o elemento em questão seja maior que o próximo é feito um swap entre eles e assim por diante, até que o primeiro elemento se encontre na posição correta, esse processo é feito N vezes, uma pra cada elemento do vetor.

### Algoritmo Bubble Sort

```python
def bubble_sort(vector):
    for i in range(0, len(vector) - 1):
        for j in range(0, len(vector) - 1):
            if vector[j] > vector[j+1]:
                vector[j], vector[j+1] = vector[j+1], vector[j]
        
    return vector
```
